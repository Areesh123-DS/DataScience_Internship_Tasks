### **Introduction and Problem Statement**
- Dataset named Iris taken from kaggle is used in this task. 
- The data set has 150 entries across 5 features and 3 classes. 
- The problem statement is to visualize the features and their relationship on Species.

### **Data Understanding and Description**
- The dataset has 6 features with species as the target column.
- The features are sepal length, sepal width, petal length, petal width,and species. 
- The species has three classes.

```python
# Loading Libraries
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as mp 
import seaborn as sns 
```

```python
data=pd.read_csv('Iris.csv') # load iris dataset
print(data.head())
print(data.info()) # Information about the data
```

```python
print("Rows,Columns: ",data.shape) # Shape of dataset
```

```python
print("Features: ",data.columns) 
```

```python
num_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
data[num_cols].hist(figsize=(14, 10), bins=20, color='skyblue', edgecolor='black')
mp.suptitle('Distribution of Features', fontsize=16,color="Black")
mp.tight_layout()
mp.show() # Visualizing the distribution of features in the dataset using histogram
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',hue="Species", data=data,edgecolor='black',palette='Set1')
mp.title('Sepal Length Vs Sepal Width',fontsize=14)
mp.xlabel('Sepal Length in cm')
mp.ylabel('Sepal Width in cm')
mp.tight_layout()
mp.show() # Visualizing the scatter plot of sepal length vs sepal width for each species
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalWidthCm',hue="Species", data=data,edgecolor='purple',palette='Set3')
mp.title('Petal Length Vs Sepal Width ',fontsize=14)
mp.xlabel('Petal Length in cm')
mp.ylabel('Sepal Width in cm')
mp.tight_layout()
mp.show() # Visualizing the scatter plot of petal length vs sepal width for each species.
```

```python
sns.boxplot(x="Species",y="SepalLengthCm",data=data,color="purple")
mp.xlabel('Species',color="Purple",fontsize=12)
mp.ylabel('Sepal Length in cm',color="Purple",fontsize=12)
# box plot showing the distribution of sepal length in cm for each species also predicting the outliers.
```

```python
sns.boxplot(x="Species",y="PetalLengthCm",data=data,color="purple")
mp.xlabel('Species',color="Purple",fontsize=12)
mp.ylabel('Petal Length in cm',color="Purple",fontsize=12)
```

### **Key Insights**
- Flowers whose petals are large are likely to have species as Iris-virginica and which have short length are iris-setosa.
- Iris-Virginica has a higher petal and sepal length than Iris-versicolor and Iris-setosa.
